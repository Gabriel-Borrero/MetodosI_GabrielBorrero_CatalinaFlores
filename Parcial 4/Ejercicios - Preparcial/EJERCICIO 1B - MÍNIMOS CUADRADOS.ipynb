```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


def line1(x, y):
    return 2*x - y - 2

def line2(x, y):
    return x + 2*y - 1

def line3(x, y):
    return x + y - 4


def total_distance(x, y):
    dist1 = line1(x, y)**2
    dist2 = line2(x, y)**2
    dist3 = line3(x, y)**2
    return np.sqrt(dist1 + dist2 + dist3)

x_values = np.arange(-5, 5, 0.01)
y_values = np.arange(-5, 5, 0.01)

X, Y = np.meshgrid(x_values, y_values)


Z = total_distance(X, Y)


min_distance = float('inf')
min_x = 0
min_y = 0

for i in range(len(x_values)):
    for j in range(len(y_values)):
        distance = total_distance(x_values[i], y_values[j])
        if distance < min_distance:
            min_distance = distance
            min_x = x_values[i]
            min_y = y_values[j]


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)

ax.scatter(min_x, min_y, min_distance, color='red', marker='o', label='Mínimo')


ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Distancia')
ax.set_title('Distancia en función de las coordenadas')


ax.legend()
plt.show()

print(f"Distancia Mínima: {min_distance}")
print(f"Coordenadas del mínimo: x = {min_x}, y = {min_y}")




```
