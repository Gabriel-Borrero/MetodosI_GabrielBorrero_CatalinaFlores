La fórmula general para la probabilidad P(n) que calcula la probabilidad de que n personas tengan fechas de cumpleaños diferentes se puede expresar de la siguiente manera:

P(n)= (365/365) × (364/365) × (363/365) × … × (365−n+1/365).     ----> Esta es la fórmula general

​


```python
import matplotlib.pyplot as plt
import numpy as np

def prob_cum_distintos(n):
    probabilidad = 1.0
    for i in range(n):
        probabilidad *= (365 - i) / 365
    return probabilidad

valores_n = np.arange(1, 81)

probabilidades = []
for n in valores_n:
    prob = prob_cum_distintos(n)
    probabilidades.append(prob)

plt.plot(valores_n, probabilidades, marker='o')
plt.xlabel('Número de personas (n)')
plt.ylabel('Probabilidad de cumpleaños diferentes')
plt.title('Probabilidad de cumpleaños diferentes en un grupo de n personas')
plt.grid(True)
plt.show()



```
